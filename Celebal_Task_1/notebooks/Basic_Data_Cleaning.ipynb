```python
import pandas as pd
```

```python
df = pd.read_csv("../data/Combined_dataset.csv")

print("Dataset Loaded Successfully")
```

```python
df.head()
```

```python
print("Shape of Dataset:")
print(df.shape)
```

```python
print("Columns in Dataset:")
print(df.columns.tolist())
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
df.info()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df["discount"] = df["discount"].fillna(df["discount"].median())
```

```python
categorical_columns = [
    "what_customers_said",
    "seller_name",
    "videos",
    "seller_information",
    "variations"
]

df[categorical_columns] = df[categorical_columns].fillna("Not Available")
```

```python
print("Remaining Missing Values:")
print(df.isnull().sum())
```

```python
print("Number of Duplicate Rows:")
print(df.duplicated().sum())
```

```python
selected_data = df[
    ["title", "category", "initial_price", "final_price"]
]

selected_data.head()
```

```python
filtered_data = df[df["rating"] >= 4]

filtered_data[
    ["title", "rating"]
].head()
```

```python
df["total_amount"] = (
    df["initial_price"] *
    df["ratings_count"]
)
```

```python
df[
    [
        "initial_price",
        "ratings_count",
        "total_amount"
    ]
].head()
```

```python
df.to_csv(
    "../output/cleaned_fashion_products.csv",
    index=False
)

print("Cleaned dataset saved successfully.")
```
